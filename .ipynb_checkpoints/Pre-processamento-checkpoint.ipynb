# Objetivo : 
### mover a coluna 'classification' para ser a ultima coluna e salvar como '0_DataFrame.csv'

```python
import os
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
dt = pd.read_csv("DUCL0001C0001.csv")
del dt['Unnamed: 0']
dt.head(2)
```

```python
data = dt
nome_colunas =list(data.columns)

dt_classification = pd.DataFrame(np.array(data['classification']),columns=['classification'])
data = data.drop(columns=['classification'],axis=1)
```

```python
data2 = pd.concat([data,dt_classification], axis=1, ignore_index=False)

data2
```

```python
#gerar o arquivo com a modificação
data2.to_csv('0_DataFrame.csv')
```
