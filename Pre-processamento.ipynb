# Objetivo : 
### mover a coluna 'classification' para ser a ultima coluna e salvar como '0_DataFrame.csv'

```python
import os
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
dataframe = pd.read_csv("DUCL0001C0001.csv")
del dataframe['Unnamed: 0']
```

```python
data = dataframe
nome_colunas =list(data.columns)

dt_classification = pd.DataFrame(np.array(data['classification']),columns=['classification'])
data.drop(columns=['classification'],axis=1,inplace=True)
```

```python
dataframe = pd.concat([data,dt_classification], axis=1, ignore_index=False)
```

```python
dataframe=dataframe[dataframe.classification!=1]
```

```python
dataframe=dataframe[dataframe.classification!=7]
```

```python
dataframe.drop(['key_point','withheld','synthetic','gps_time','user_data','X','point_source_id'],axis=1,inplace=True)
dataframe
```

```python
dataframe=dataframe.apply(lambda x: x.replace(True,int(1)))
```

```python
dataframe=dataframe.apply(lambda x: x.replace(False,int(0)))
```

```python
#gerar o arquivo com a modificação
dataframe.to_csv('0_DataFrame.csv')
```
